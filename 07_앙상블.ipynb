```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd

cancer = load_breast_cancer()

data_df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
data_df['target'] = cancer.target
data_df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split(
    data_df.drop("target", axis=1),
    data_df['target'],
    random_state=42
)
```

# Voting Classifier
image.png

```python
from sklearn.ensemble import VotingClassifier # VotingCalssifierëŠ” ëª¨ë¸ë“¤ì„ ëª¨ì•„ì£¼ëŠ” ì—­í• ë§Œ í•¨.

# VotingCalssifierì— ë“¤ì–´ê°ˆ ëª¨ë¸
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import LinearSVC
```

## Hard Voting

```python
# ì•™ìƒë¸”ì— ì°¸ì—¬í•  ëª¨ë¸ ê°ì²´ ìƒì„±
lr_clf = LogisticRegression()
knn_clf = KNeighborsClassifier(n_neighbors=3)
svc = LinearSVC()

estimators = [
    ('LR', lr_clf), # lr_clfë¥¼ "LR"ì´ë¼ëŠ” ì´ë¦„ìœ¼ë¡œ ì„¤ì •
    ('KNN', knn_clf),
    ('SVC', svc)

]

vo_clf = VotingClassifier(
    estimators = estimators,
    voting = 'hard' # í•˜ë“œë³´íŒ… = ë‹¤ìˆ˜ê²°

)
```

```python
vo_clf.fit(X_train, y_train)
```

```python
# Hard Voting ë°©ì‹ì—ì„œ ì„±ëŠ¥í‰ê°€
from sklearn.metrics import accuracy_score
pred = vo_clf.predict(X_test)
accuracy_score(y_test, pred)
```

```python
# ê°œë³„ ëª¨ë¸ ê°ê°ì˜ ì„±ëŠ¥ í™•ì¸
for estimator in vo_clf.estimators_:
  pred = estimator.predict(X_test)
  class_name = estimator.__class__.__name__

  print("{} ì •í™•ë„ : {:.4f}".format(class_name, accuracy_score(y_test, pred)))
```

## Soft Voting

```python
# ì•™ìƒë¸”ì— ì°¸ì—¬í•  ëª¨ë¸ ê°ì²´ ìƒì„±
lr_clf = LogisticRegression()
knn_clf = KNeighborsClassifier(n_neighbors=3)
svc = LinearSVC()

estimators = [
    ('LR', lr_clf),
    ('KNN', knn_clf),
   #('SVC', svc) svc ëª¨ë¸ì€ predict_probaê°€ ì—†ê¸° ë•Œë¬¸ì— soft votingì— ì í•©í•˜ì§€ ì•Šë‹¤.
   #soft votingì€ í™•ë¥  ê°’ì´ í•„ìš”í•œ LinearSVC()ë¡œ, predict_probaê°€ ì—†ëŠ” ëª¨ë¸ì„ ì‚¬ìš©í•  ìˆ˜ ì—†ë‹¤.
]

vo_clf = VotingClassifier(
    estimators=estimators,
    voting='soft'
)
```

```python
# Hard Voting ë°©ì‹ì—ì„œ ì„±ëŠ¥í‰ê°€
from sklearn.metrics import accuracy_score
pred = vo_clf.predict(X_test)
accuracy_score(y_test, pred)
```

```python
# ê°œë³„ ëª¨ë¸ ê°ê°ì˜ ì„±ëŠ¥ í™•ì¸
for estimator in vo_clf.estimators_:
  pred = estimator.predict(X_test)
  class_name = estimator.__class__.__name__

  print("{} ì •í™•ë„ : {:.4f}".format(class_name, accuracy_score(y_test, pred)))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=100, noise=0.25, random_state=3)
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)
```

```python
forest_clf = RandomForestClassifier(n_estimators=5, random_state=2)
forest_clf.fit(X_train, y_train)
```

```python
import numpy as np

# Classifierì˜ Decision Boundaryë¥¼ ì‹œê°í™” í•˜ëŠ” í•¨ìˆ˜
def visualize_boundary(model, X, y, ax):

    # í•™ìŠµ ë°ì´íƒ€ scatter plotìœ¼ë¡œ ë‚˜íƒ€ë‚´ê¸°
    ax.scatter(X[:, 0], X[:, 1], c=y, s=25, cmap='rainbow', edgecolor='k',
               clim=(y.min(), y.max()), zorder=3)
    ax.axis('tight')
    ax.axis('off')
    xlim_start , xlim_end = ax.get_xlim()
    ylim_start , ylim_end = ax.get_ylim()

    # í˜¸ì¶œ íŒŒë¼ë¯¸í„°ë¡œ ë“¤ì–´ì˜¨ training ë°ì´íƒ€ë¡œ model í•™ìŠµ .
    model.fit(X, y)

    # meshgrid í˜•íƒœì¸ ëª¨ë“  ì¢Œí‘œê°’ìœ¼ë¡œ ì˜ˆì¸¡ ìˆ˜í–‰.
    xx, yy = np.meshgrid(np.linspace(xlim_start,xlim_end, num=200),np.linspace(ylim_start,ylim_end, num=200))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

    # contourf() ë¥¼ ì´ìš©í•˜ì—¬ class boundary ë¥¼ visualization ìˆ˜í–‰.
    n_classes = len(np.unique(y))
    contours = ax.contourf(xx, yy, Z, alpha=0.3,
                           levels=np.arange(n_classes + 1) - 0.5,
                           cmap='rainbow', clim=(y.min(), y.max()),
                           zorder=1)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 3, figsize=(20, 10))

for i, (ax, tree) in enumerate(zip(axes.ravel(), forest_clf.estimators_)):
  ax.set_title("Tree {}".format(i+1))
  visualize_boundary(tree, X, y, ax)

# ì œì¼ ë§ˆì§€ë§‰ ax ì—ëŠ” RandomForest
axes.ravel()[-1].set_title("RandomForest")
visualize_boundary(forest_clf, X, y, axes.ravel()[-1])

plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split(
    data_df.drop("target", axis=1),
    data_df['target'],
    random_state=42
)
```

```python
from sklearn.model_selection import GridSearchCV

params = {
    "n_estimators" : [10, 100] # 10ê°œì¼ ë•Œì™€, 100ê°œì¼ ë•Œ
}

rf_grid_search = GridSearchCV(
    RandomForestClassifier(random_state=42),
    param_grid = params,
    scoring='accuracy',
    cv=5
)
```

```python
rf_grid_search.fit(X_train, y_train)
```

```python
# GridSearchCVê°ì²´ì˜ cv_results_ ì†ì„±ì„ DataFrameìœ¼ë¡œ ìƒì„±.
cv_results_df = pd.DataFrame(rf_grid_search.cv_results_)

# max_depth,min_samples_split íŒŒë¼ë¯¸í„° ê°’ê³¼ ê·¸ë•Œì˜ í…ŒìŠ¤íŠ¸(Evaluation)ì…‹, í•™ìŠµ ë°ì´í„° ì…‹ì˜ ì •í™•ë„ ìˆ˜ì¹˜ ì¶”ì¶œ
cv_results_df[['param_n_estimators', 'mean_test_score']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
def plot_feature_importance(model, columns, limit=None):

  ftr_importances_values = model.feature_importances_
  # Top ì¤‘ìš”ë„ë¡œ ì •ë ¬ì„ ì‰½ê²Œ í•˜ê³ , ì‹œë³¸(Seaborn)ì˜ ë§‰ëŒ€ê·¸ëž˜í”„ë¡œ ì‰½ê²Œ í‘œí˜„í•˜ê¸° ìœ„í•´ Seriesë³€í™˜
  ftr_importances = pd.Series(ftr_importances_values, index=columns  )
  # ì¤‘ìš”ë„ê°’ ìˆœìœ¼ë¡œ Seriesë¥¼ ì •ë ¬
  ftr_top = ftr_importances.sort_values(ascending=False)[:limit]
  plt.figure(figsize=(8,6))
  plt.title(f'Feature importances Top {len(columns) if limit == None else limit}')
  sns.barplot(x=ftr_top , y = ftr_top.index)
  plt.show()

plot_feature_importance(rf_grid_search.best_estimator_, X_train.columns)
```

```python
# ì˜ì‚¬ê²°ì • íŠ¸ë¦¬
from sklearn.tree import DecisionTreeClassifier

from sklearn.model_selection import GridSearchCV

params = {
    'max_depth' : [8, 12, 16 ,20],
    'min_samples_split' : [16, 24],
}

dt_grid_search = GridSearchCV(
    DecisionTreeClassifier(random_state=42),
    param_grid=params,
    cv=5,
    scoring='accuracy'
)

dt_grid_search.fit(X_train, y_train)
```
